I'll rename some of the columns to be easier to work with

```python
df.rename(columns = {"Steps (steps)": "Steps", "Flights Climbed (floors)": "Floors Climbed", "Stand Hours (hr)": "Stand Hours", "Stand Minutes (min)": "Stand minutes"}, inplace = True)
```

```python
df.head()
```

I'll convert the 'Date' column from a string to a DateTime

```python
df['Date'] = pd.to_datetime(df['Date'],dayfirst=True)
```

```python
df['Date'].head()
```

```python
df['Day of Week'] = df['Date'].apply(lambda time: time.dayofweek)
df['Month'] = df['Date'].apply(lambda time: time.month)
df['Year'] = df['Date'].apply(lambda time: time.year)
```

```python
df.head()
```

And move the columns to have the dates/times all together

```python
col = df.pop('Day of Week')
df.insert(1,'Day of Week',col)
```

```python
col = df.pop('Month')
df.insert(2,'Month',col)
```

```python
col = df.pop('Year')
df.insert(3,'Year',col)
```

```python
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
df.head()
```
